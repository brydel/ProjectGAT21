```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
from statsmodels.tsa.seasonal import seasonal_decompose
import warnings
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from math import sqrt
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from IPython.display import FileLink
```

```python
# Compter le nombre de valeurs manquantes par colonne
missing_values_count = df.isnull().sum()
print(missing_values_count)
```

```python
df.head(3)
```

```python
# Supprimer les lignes où la température est manquante
df = df.dropna(subset=['Temp (°C)'])
```

```python
# Remplacer les valeurs manquantes dans 'Temp (°C)' par la moyenne
df['Temp (°C)'].fillna(df['Temp (°C)'].mean(), inplace=True)
```

```python
df['Temp (°C)'] = pd.to_numeric(df['Temp (°C)'],  errors = 'coerce')
df['Visibility (km)'] = pd.to_numeric(df['Visibility (km)'],  errors = 'coerce')
df['Stn Press (kPa)'] = pd.to_numeric(df['Stn Press (kPa)'],  errors = 'coerce')
```

```python
#remplacement des valeur manquantes
```

```python
for col in df.select_dtypes(include = ['float', 'int']):
    df[col] = df[col].fillna(df[col].mean())

#remplisage des valeur NAN pour chaque chauque collones specifique 
df['Temp (°C)'] = df['Temp (°C)'].fillna(df['Temp (°C)'].mean())
df['Visibility (km)'] = df['Visibility (km)'].fillna(df['Visibility (km)'].mean())
df['Stn Press (kPa)'] = df['Stn Press (kPa)'].fillna(df['Stn Press (kPa)'].mean())
```

```python
#je vais supprimer les colonnes avec plus de 50% de valeur manquantes 
df.dropna(axis =1,thresh=len(df)*0.5, inplace=True)
```

```python
df.isna().sum()
```

```python
df.head(3)
```

```python
# calcul des tendances des temperatures en utilisant les donner climatiques 
```

```python
#convertir les date en format datetime
df['Date/Time (LST)'] = pd.to_datetime(df['Year'].astype(str) + '-' +
                                        df['Month'].astype(str).str.zfill(2) + '-' +
                                        df['Day'].astype(str).str.zfill(2) + ' ' +
                                        df['Time (LST)'])
df.set_index('Date/Time (LST)', inplace=True)
```

```python
df.head(3)
```

```python
# Vérification des types de données pour les colonnes numériques
df['Longitude (x)'] = pd.to_numeric(df['Longitude (x)'], errors='coerce')
df['Latitude (y)'] = pd.to_numeric(df['Latitude (y)'], errors='coerce')


```

```python
# Traitement des valeurs manquantes pour Temp (°C) en remplaçant par la moyenne
df.loc[:, 'Temp (°C)'] = df['Temp (°C)'].fillna(df['Temp (°C)'].mean())

```

```python
#calcul de la moyenne des temperature annuelle et mensuel 
```

```python

# Traitement des autres mesures météorologiques de la même manière
df.loc[:, 'Dew Point Temp (°C)'] = df['Dew Point Temp (°C)'].fillna(df['Dew Point Temp (°C)'].mean())
df.loc[:, 'Rel Hum (%)'] = df['Rel Hum (%)'].fillna(df['Rel Hum (%)'].mean())
df.loc[:, 'Wind Dir (10s deg)'] = df['Wind Dir (10s deg)'].fillna(df['Wind Dir (10s deg)'].mode()[0])
df.loc[:, 'Wind Spd (km/h)'] = df['Wind Spd (km/h)'].fillna(df['Wind Spd (km/h)'].mean())
df.loc[:, 'Visibility (km)'] = df['Visibility (km)'].fillna(df['Visibility (km)'].mean())
df.loc[:, 'Stn Press (kPa)'] = df['Stn Press (kPa)'].fillna(df['Stn Press (kPa)'].mean())
```

```python
#Analyse de la serie Temporelle
#j'alanyse de la serie temporelles ici vas m'aider a révéler des tendances, des saisonnalités et des cycles dans les données qui ne sont pas apparents dans les graphiques de dispersion.
result = seasonal_decompose(df['Temp (°C)'], model='additive')
plt.figure(figsize=(10, 4))
result.plot()
plt.tight_layout() 
plt.show()
```

```python
#Décomposition de Série Temporelle
#j'uilisez la décomposition saisonnière pour identifier les tendances et les schémas saisonniers dans les données de température.
#biblio  statsmodels.tsa.seasonal, seasonal_decompose
plt.figure(figsize=(10, 4))
plt.plot(result.resid)
plt.title('Résidus de la Décomposition Saisonnière')
plt.xlabel('Date')
plt.ylabel('Résidus')
plt.tight_layout()  # Ajuste le tracé pour qu'il s'adapte bien à la figure
plt.show()
```

```python
#distribution:
plt.figure(figsize=(8, 4))
sns.histplot(result.resid.dropna(), kde=True)  # Retirez les NaN avant de tracer
plt.title('Distribution des Résidus')
plt.xlabel('Résidus')
plt.show()
```

```python
import pandas as pd
import openpyxl
from openpyxl.styles import Font, Color, PatternFill

def detect_and_replace_suspect_values(df):
    # Identifier les colonnes avec des données numériques
    colonnes_numeriques = df.select_dtypes(include=['number']).columns
    
    # Remplacer les valeurs nulles par la moyenne dans chaque colonne numérique
    for colonne in colonnes_numeriques:
        if df[colonne].isnull().any():  # Vérifier s'il y a des valeurs NaN
            valeur_moyenne = df[colonne].mean()
            df[colonne].fillna(valeur_moyenne, inplace=True)
    
    return df

# Chemin vers le fichier CSV
chemin_fichier_csv = 'c:\\users\\brydel\\desktop\\projetgat21\\gatineau2.csv'

# Lire le fichier CSV
df = pd.read_csv(chemin_fichier_csv)

# Détecter et remplacer les valeurs suspectes
df = detect_and_replace_suspect_values(df)

# Exporter en Excel avec openpyxl
path = 'output_styled.xlsx'
df.to_excel(path, index=False, engine='openpyxl')

# Ouvrir le fichier Excel pour appliquer le style
wb = openpyxl.load_workbook(path)
ws = wb.active

red_fill = PatternFill(start_color='FFFF0000',
                        end_color='FFFF0000',
                        fill_type='solid')

for row in ws.iter_rows(min_row=2, max_row=ws.max_row, max_col=ws.max_column):
    for cell in row:
        if cell.value and isinstance(cell.value, (int, float)) and cell.value > 10:
            cell.fill = red_fill

wb.save(path)
print("DataFrame exporté avec succès avec les valeurs > 10 en rouge.")
```
